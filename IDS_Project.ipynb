```python
median_glucose=df['Glucose'].median()
df['Glucose']=df['Glucose'].replace(0,median_glucose)
df.describe()
```

```python
median_bloodpressure=df['BloodPressure'].median()
df['BloodPressure']=df['BloodPressure'].replace(0,median_bloodpressure)
df.describe()
```

```python
median_skinthickness=df['SkinThickness'].median()
df['SkinThickness']=df['SkinThickness'].replace(0,median_skinthickness)
df.describe()
```

```python
median_insulin=df['Insulin'].median()
df['Insulin']=df['Insulin'].replace(0,median_insulin)
df.describe()
```

```python
median_bmi=df['BMI'].median()
df['BMI']=df['BMI'].replace(0,median_bmi)
df.describe()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df, width=0.5)
plt.title('Boxplots for Each Column')
plt.show()
```

```python
sns.pairplot(df, hue='Outcome')
plt.show()
#sto su slucajevi blize obliku kose ravne linije, to je njihova korelacija veca
```

```python
class_counts = df.groupby('Outcome').size() 
print("Class breakdown of the data:")
print(class_counts)
```

```python
df.groupby('Outcome').mean().T.plot(figsize=(10,5))
```

```python
#1 tj narancasti su oni koji imaju diabetes, dok su plavi oni koji nemaju
```

```python
categories = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

plt.figure(figsize=(10, 18))

for category in categories:
    plt.subplot(3, 3, categories.index(category) +1)
    sns.barplot(x='Outcome', y=category, data=df, color='pink' )
    plt.title(f'{category} by Outcome')

plt.tight_layout()
plt.show()
```
